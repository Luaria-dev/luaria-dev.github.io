```python
from PIL import Image
import os

def separar_gif_en_webp(gif_entrada, carpeta_salida="frames_webp", nombre_base="frame"):
    """
    Separa un GIF animado en imágenes WebP.

    gif_entrada: ruta del GIF original
    carpeta_salida: carpeta donde se guardarán los frames
    nombre_base: nombre base de cada imagen generada
    """

    os.makedirs(carpeta_salida, exist_ok=True)

    gif = Image.open(gif_entrada)

    total_frames = getattr(gif, "n_frames", 1)

    for i in range(total_frames):
        gif.seek(i)

        frame = gif.convert("RGBA")

        ruta_salida = os.path.join(
            carpeta_salida,
            f"{nombre_base}_{i:03d}.webp"
        )

        frame.save(
            ruta_salida,
            "WEBP",
            lossless=True,
            quality=100
        )

        print(f"Frame guardado: {ruta_salida}")

    print(f"\nListo. Se exportaron {total_frames} frames en '{carpeta_salida}'.")


# Ejemplo de uso
separar_gif_en_webp(
    gif_entrada="recruiter_80_100.gif",
    carpeta_salida="recruiter_80_100",
    nombre_base="recruiter_80_100"
)
```
